```python
def is_prime(num):
    """ Fungsi untuk memeriksa apakah sebuah bilangan prima """
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

n = 19
start  = 1000 * n
end = 1000 * (n + 1)

print(f"Bilangan prima antara {start} dan {end} adalah:")
for num in range(start, end):
    if is_prime(num):
        print(num)
```
